## Feature Engineering

```
# Encode categorical variables.
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()

enc_df = df
enc_df['businessTravel'] = ord_enc.fit_transform(df[['BusinessTravel']])
enc_df['department'] = ord_enc.fit_transform(df[['Department']])
enc_df['educationField'] = ord_enc.fit_transform(df[['EducationField']])
enc_df['gender'] = ord_enc.fit_transform(df[['Gender']])
enc_df['jobRole'] = ord_enc.fit_transform(df[['JobRole']])
enc_df['maritalStatus'] = ord_enc.fit_transform(df[['MaritalStatus']])
```

```
# Drop original categorical variables.
enc_df = enc_df.drop(['BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus'], axis=1)
```

```
enc_df.head()
```

```
# Scale the data and define x and y.
x = enc_df.drop('Attrition', axis=1)
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(x)
y = enc_df.Attrition
```

## Deep Learning Model

```
# Split the data into train and test sets.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```
X_train.shape
```

```
import tensorflow as tf
```

```
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=200, activation='relu', input_shape=(30, )))
model.add(tf.keras.layers.Dense(units=100, activation='relu'))
model.add(tf.keras.layers.Dense(units=50, activation='relu'))
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```
model.summary()
```

```
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
# Early stopping.
from keras.callbacks import EarlyStopping
monitor = EarlyStopping(patience=3)
```

```
epochs_hist = model.fit(X_train, y_train, validation_split=0.3, epochs = 50, batch_size = 30, callbacks=[monitor])
```

```
epochs_hist.history.keys()
```

```
plt.plot(epochs_hist.history['loss'], label='train')
plt.plot(epochs_hist.history['val_loss'], label='test')
plt.xlabel('Epoch')
plt.title('Training vs. validation loss')
plt.legend()
plt.show()
```

```
plt.plot(epochs_hist.history['accuracy'], label='train')
plt.plot(epochs_hist.history['val_accuracy'], label='test')
plt.title('Model Accuracy Training vs. Testing')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```
#p_pred = model.predict(X_test)
#p_pred = p_pred.flatten()
#y_pred = np.where(p_pred > 0.5, 1, 0)    

y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)
```

```
# Confusion Matrix.
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

```
print(classification_report(y_test, y_pred))
```
